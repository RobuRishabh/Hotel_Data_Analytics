# Step.2 : Data Cleaning

```python
df_bookings.describe()
```

```python
# let's check for any negative values
df_bookings[df_bookings.no_guests<=0]
```

```python
df_bookings.shape
```

```python
# now let's get rid of all the negative no.of guests
df_bookings = df_bookings[df_bookings.no_guests>=0]   #means store only zero and positive values in the df again
df_bookings.shape
```

```python
df_bookings.revenue_generated.min(),df_bookings.revenue_generated.max() 
# Here the max revenue is too much which is not possible
```

```python
avg, std = df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.std()
avg, std
```

```python
higher_limit = avg + 3*std
higher_limit
```

```python
low_limit = avg - 3*std
low_limit
```

```python
df_bookings[df_bookings.revenue_generated > higher_limit]
```

```python
# Now let's remove all those values whose revenue generated are insane
df_bookings = df_bookings[df_bookings.revenue_generated < higher_limit]
df_bookings.shape
```

# Step.3 : Data Transformation

```python
df_agg_bookings.head()
```

```python
# Occupancy Percentage = successful_bookings / capacity
# lets add a new column which shows occupancy percentage
df_agg_bookings["occ_pct"] = df_agg_bookings["successful_bookings"]/df_agg_bookings["capacity"]
```

```python
df_agg_bookings.head()
```

See now we have occupancy percentage columns also

```python
# apply() if we have a column and we want to apply transformation on that column and create a new column we use apply()
df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(lambda x : round(x*100, 2))
df_agg_bookings.head()
```
